```python
df['TotalChildren'] = df['Kidhome'] + df['Teenhome']
df
```

```python
df['Total_Purchases'] = (
    df['NumWebPurchases'] +
    df['NumCatalogPurchases'] +
    df['NumStorePurchases']
)
print(df[['ID', 'Total_Purchases']].head())
```

```python
latest_date = df['Dt_Customer'].max()
df['Customer_Tenure_Days'] = (
    latest_date - df['Dt_Customer']
).dt.days
print(df[['ID', 'Customer_Tenure_Days']].head())
```

```python
df['Customer_Year'] = df['Dt_Customer'].dt.year
df['Customer_Month'] = df['Dt_Customer'].dt.month
df['Customer_Day'] = df['Dt_Customer'].dt.day
print(df[['ID', 'Customer_Year', 'Customer_Month', 'Customer_Day']].head())
```

```python
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
print(lower, upper)
```

```python
df['Income'] = np.where(
    df['Income'] > upper,
    upper,
    df['Income']
)
df['Income'] = np.where(
    df['Income'] < lower,
    lower,
    df['Income']
)
print(df['Income'].describe())
```

```python
df = df[df['Age'] <= 100]
print(df['Age'].describe())
```

```python
numeric_cols = df.select_dtypes(include=np.number)
print(
    numeric_cols.skew().sort_values(
        ascending=False
    )
)
```

```python
spending_cols = [
    'Income',
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]
for col in spending_cols:
    df[col] = np.log1p(df[col])
    print(df[col].describe())
```

```python
print(df[spending_cols].skew())
```

```python
df['Marital_Status'].unique()
print(df['Marital_Status'].value_counts())
```

```python
# Rename Education categories

df['Education'] = df['Education'].replace({
    'Basic': 'Undergraduate',
    '2n Cycle': 'Undergraduate',
    'Graduation': 'Graduate',
    'Master': 'Postgraduate',
    'PhD': 'Postgraduate'
})

# Display counts of each category
print(df['Education'].value_counts())
```

```python

# Remove extra spaces in column names
df.columns = df.columns.str.strip()
# Remove duplicates
df = df.drop_duplicates()
# Fill missing values in numeric columns with median
numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())
# Create Age column
if 'Year_Birth' in df.columns:
    df['Age'] = 2026 - df['Year_Birth']
# Create Total Spending
spending_cols = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]
existing_spending = [col for col in spending_cols if col in df.columns]
if existing_spending:
    df['TotalSpending'] = df[existing_spending].sum(axis=1)
# Create Total Purchases
purchase_cols = [
    'NumWebPurchases',
    'NumCatalogPurchases',
    'NumStorePurchases'
]
existing_purchase = [col for col in purchase_cols if col in df.columns]
if existing_purchase:
    df['TotalPurchases'] = df[existing_purchase].sum(axis=1)
# Save cleaned dataset in Colab Files
df.to_csv("cleaned_marketing_data.csv", index=False)
print("Data Cleaning Completed Successfully")
print("Final Shape:", df.shape)
print("File saved as: cleaned_marketing_data.csv")
```
